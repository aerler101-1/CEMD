```python
def normalize_studentid(df):
    for col in df.columns:
        if col.lower() == "studentid" or col.upper() == "ID":
            df = df.rename(columns={col: "StudentID"})
    return df

map_df = normalize_studentid(map_df)
enrollments_df = normalize_studentid(enrollments_df)
students_df = normalize_studentid(students_df)
attendance_df = normalize_studentid(attendance_df)

# === Attendance Logic ===
present_codes = ["P", "UT", "UA"]
tardy_codes = ["UT", "ET"]

attendance_df["is_present"] = attendance_df["AttendanceCode"].isin(present_codes)
attendance_df["is_tardy"] = attendance_df["AttendanceCode"].isin(tardy_codes)

# === Attendance Rate Summary ===
attendance_summary = (
    attendance_df.groupby("StudentID")
    .agg(
        total_days=("EventDate", "nunique"),
        present_days=("is_present", "sum")
    )
    .assign(attendance_rate=lambda df: df["present_days"] / df["total_days"])
    .reset_index()
)

# === Tardy Rate Summary ===
tardy_summary = (
    attendance_df.groupby("StudentID")
    .agg(
        present_days=("is_present", "sum"),
        tardy_days=("is_tardy", "sum")
    )
    .assign(tardy_rate=lambda df: df["tardy_days"] / df["present_days"])
    .reset_index()
)

# === Total Days (Again for clarity) ===
total_days_df = (
    attendance_df.groupby("StudentID")["EventDate"]
    .nunique()
    .reset_index(name="total_days")
)

# === Most Common SchoolID per Student ===
schoolid_per_student = (
    attendance_df.groupby("StudentID")["schoolid"]
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else pd.NA)
    .reset_index()
)

# === Final Merge: The Normed Student Table ===
student = (
    attendance_summary[["StudentID", "attendance_rate"]]
    .merge(tardy_summary[["StudentID", "tardy_rate"]], on="StudentID", how="outer")
    .merge(total_days_df, on="StudentID", how="outer")
    .merge(schoolid_per_student, on="StudentID", how="left")
)

student.set_index("StudentID", inplace=True)

# === Preview result ===
student.head()
```

```python
filtered_enrollments = enrollments_df[enrollments_df["StudentID"].isin(student.index)]

# Step 3: Define helper function to extract up to 2 TEACHERIDs as strings
def extract_teacher_ids(df, subject_code):
    subject_df = df[df["COURSE_NUMBER"].str.contains(subject_code, na=False)].copy()
    subject_df["TEACHERID"] = subject_df["TEACHERID"].astype(str)  # ensure string
    
    teacher_lookup = (
        subject_df.groupby("StudentID")["TEACHERID"]
        .apply(lambda x: x.iloc[:2].tolist() if len(x) >= 1 else [None, None])
        .apply(lambda lst: lst + [None] * (2 - len(lst)))  # pad to 2 items
        .apply(pd.Series)
        .rename(columns={0: f"{subject_code.lower()}_teacher_1", 1: f"{subject_code.lower()}_teacher_2"})
    )
    
    return teacher_lookup

# Step 4: Extract teachers for Math (MAT) and ELA
math_teachers = extract_teacher_ids(filtered_enrollments, "MAT")
ela_teachers = extract_teacher_ids(filtered_enrollments, "ELA")

# Step 5: Merge into student DataFrame
student = student.join(math_teachers, how="left")
student = student.join(ela_teachers, how="left")

# Preview
student.head()
```

```python
map_df["StudentID"] = map_df["StudentID"].astype(int)
student.reset_index(inplace=True)
student["StudentID"] = student["StudentID"].astype(int)

map_df = map_df[map_df["StudentID"].isin(student["StudentID"])]

# Clean term/subject strings
map_df["map_term"] = map_df["map_term"].str.strip().str.capitalize()
map_df["subject"] = map_df["subject"].str.strip().str.capitalize()

# Subset columns for pivoting
map_df_subset = map_df[[
    "StudentID", "year", "map_term", "subject",
    "testdurationinminutes", "testritscore", "testpercentile", "typicalfalltofallgrowth"
]]

# Define pivot function
def pivot_map_feature(df, value_col, prefix):
    df_copy = df.copy()
    df_copy["col_key"] = (
        prefix + "_" +
        df_copy["year"].astype(str) + "_" +
        df_copy["map_term"] + "_" +
        df_copy["subject"].str.lower().str.replace(" ", "_")
    )
    return df_copy.pivot(index="StudentID", columns="col_key", values=value_col)

# Pivot features
duration_wide   = pivot_map_feature(map_df_subset, "testdurationinminutes", "duration")
rit_wide        = pivot_map_feature(map_df_subset, "testritscore", "rit")
percentile_wide = pivot_map_feature(map_df_subset, "testpercentile", "percentile")

# Restrict FTF to Fall 2015 only
ftf_filtered = map_df_subset[
    (map_df_subset["year"] == 2015) & (map_df_subset["map_term"] == "Fall")
]
ftf_wide = pivot_map_feature(ftf_filtered, "typicalfalltofallgrowth", "ftf")

# Combine wide-format tables
map_wide = duration_wide.join(rit_wide).join(percentile_wide).join(ftf_wide).reset_index()

# Merge with student table
student = student.merge(map_wide, on="StudentID", how="left").set_index("StudentID")

# Preview
student.head()
```

```python
import pandas as pd

# Load your data
df = pd.read_csv("student_table_enriched.csv")

# Melt teacher subject columns into long format
teacher_long = pd.melt(
    df,
    id_vars=["StudentID"],
    value_vars=["mat_teacher_1", "mat_teacher_2", "ela_teacher_1", "ela_teacher_2"],
    var_name="subject_slot",
    value_name="teacher_id"
)

# Drop rows with null teacher IDs
teacher_long = teacher_long.dropna(subset=["teacher_id"]).drop_duplicates()

# Map subject slot to subject label
teacher_long["subject"] = teacher_long["subject_slot"].apply(lambda x: "Math" if "mat_" in x else "ELA")

# Count how many subjects each teacher taught
subject_counts = (
    teacher_long.groupby("teacher_id")["subject"]
    .nunique()
    .reset_index(name="num_subjects")
    .query("num_subjects > 1")
)

# Show teachers and which subjects they taught
teachers_multi_subjects = teacher_long[teacher_long["teacher_id"].isin(subject_counts["teacher_id"])]
teachers_multi_subjects = teachers_multi_subjects[["teacher_id", "subject"]].drop_duplicates().sort_values(["teacher_id", "subject"])

# Display result
teachers_multi_subjects
```
